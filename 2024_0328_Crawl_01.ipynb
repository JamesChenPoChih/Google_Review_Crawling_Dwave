```python
from langdetect import detect
from langdetect.lang_detect_exception import LangDetectException

# 定义一个函数来检测语言
def detect_language(text):
    try:
        # 检测语言
        language = detect(text)
        return language
    except LangDetectException as e:
        print("Error:", e)
        return None

# 为每个评论检测语言并将结果存储在新列中
data['language'] = data['content'].apply(lambda x: detect_language(str(x)))

# 显示新的语言列
print(data['language'])
```
